```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 6))
%matplotlib inline
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'. format(col,pct_missing))
```

```python
df.dropna(inplace= True)
```

```python
df.isnull().count()
```

```python
df['budget'] = df['budget'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head(20)
```

```python
df['released_date'] = df.released.str.split('(', expand=True)[0]
df['released_date'] = pd.to_datetime(df.released_date)
df['year'] = df.released_date.dt.year
df['released_country'] = df.released.str.split('(', expand=True)[1].str.replace(r'[^a-zA-Z\d\s:]', '')
df = df.drop(['released','country'], axis =1)
```

```python
df
```

```python
df = df.sort_values(by=['gross'],inplace=False,ascending= False)
```

```python
pd.set_option('display.max_rows', None)
```

# we can check out correlation between 
#bugdet vs gross

```python
plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earning')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earning')
plt.show()
```

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
df.corr(method= 'pearson')
```

```python
Corr_Mat = df.corr(method= 'pearson')
sns.heatmap(Corr_Mat,annot= True)
plt.title('Correlation For Numeric Values')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
df_numerize = df
for col_name in df_numerize.columns:
    if(df_numerize[col_name].dtype == 'object'):
        df_numerize[col_name] = df_numerize[col_name].astype('category')
        df_numerize[col_name] = df_numerize[col_name].cat.codes
    df_numerize.head()
```

```python
Corr_Mat = df_numerize.corr(method= 'pearson')
fig, ax = plt.subplots(figsize=(14, 6))
sns.heatmap(Corr_Mat,annot= True, ax=ax)
plt.title('Correlation For Numeric Values')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
correlation_mat = df_numerize.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
